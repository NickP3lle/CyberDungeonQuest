```python
# Encryption function
def encrypt(message, e, n):
    ciphertext = pow(message, e, n)
    return ciphertext
```

We have found the lenght of the flag which is 24 + CCIT{...} = 30. Brute forcing the flag would be too expensive on time. But we can see that the first part of what we encrypted is the same as the ciphertext. Using this information we can find the flag by checking characters if they match the ciphertext. If they do we can add them to the flag.

```python
# Function to count the number of matching digits between two numbers
def count_matching_digits(num1, num2):
    str_num1 = str(num1)
    str_num2 = str(num2)

    # The lenght in our case is the same for both numbers
    lenght = len(str_num1)

    matching_digits = 0

    # Iterate over digits until a mismatch is found
    for i in range(lenght):
        if str_num1[i] == str_num2[i]:
            matching_digits += 1
        else:
            break

    return matching_digits
```

Trying an iterative approach doesn't seem to work because, working character by character, we get multiple possible characters for each position and their number is not defined a priori. Using a recursive approach seems to be the best way to solve this problem. The following function does just that.
Every time we find a character that matches we need to see if there are other characters that match the same position and continue the search for the following positions (opt1 anf opt2 in the function). If we find a character that doesn't match we need to continue the search for the same position.
A character that matches the ciphertext is added to the flag and the search continues for the next position. Then the two possibilities are compared to find the longest. Of course the longest one in the end is the flag.

```python
def find_next_possible_char(flag, custom_range, e, n, c, previus_match, i=0, j=0, deviation=0):
    # Note: 24 is the length that we have found before
	if i == 24:
		plaintext = "CCIT{" + flag + "}"
		message = int.from_bytes(plaintext.encode(), 'big')
		enc = encrypt(message, e, n)
		
		new_match = count_matching_digits(c, enc)
		
		if new_match == len(str(c)):
			return flag

	if j == len(custom_range):
		return ""

	plaintext = "CCIT{" + flag + chr(custom_range[j]) + "a" * (23 - i) + "}"
	message = int.from_bytes(plaintext.encode(), 'big')
	enc = encrypt(message, e, n)

	new_match = count_matching_digits(c, enc)
	difference = new_match - previus_match + deviation


	if (difference > 0):
		opt1 = find_next_possible_char(flag + chr(custom_range[j]), custom_range, e, n, c, new_match, i + 1, 0, 0)
		opt2 = find_next_possible_char(flag, custom_range, e, n, c, previus_match, i, j + 1, deviation)

		if len(opt1) > len(opt2):
			return opt1
		else:
			return opt2
	else:
		deviation = 0
		if difference < 0:
			deviation = abs(difference)

		return find_next_possible_char(flag, custom_range, e, n, c, previus_match, i, j + 1, deviation)
```
